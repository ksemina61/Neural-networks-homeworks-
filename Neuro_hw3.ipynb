```python
def preprocess(text):
    text = re.sub(r'https?://\S+|www\.\S+', '', text) # удалю ссылки с помощью регулярок
    text = text.lower() # все к нижнему регистру
    text = ''.join([char for char in text if char not in string.punctuation]) #удаляются все знаки препинания
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words] # удаляются стоп-слова
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens] #леммантизация
    return tokens
```

```python
# функция для предобработки текста, которая получает сырой текст и возвращает список токенов
data['tokens'] = data['case_text_title'].apply(preprocess)
```

```python
class LegalDataset(Dataset):

    def __init__(self, data, word2id, DEVICE):
        self.data = data['case_text_title'].values # x хранятся в тексте
        self.targets = data['case_outcome'].values
        self.word2id = word2id # словарь
        self.length = data.shape[0] # длинв
        self.le = LabelEncoder()
        self.targets = self.le.fit_transform(self.targets)
        self.device = DEVICE

    def __len__(self): #это обязательный метод, он должен уметь считать длину датасета
        return self.length

    def __getitem__(self, index):
      text = self.data[index]
      tokens = preprocess(text)
      ids = torch.LongTensor([self.word2id[token] for token in tokens if token in self.word2id])
      return ids, self.targets[index]


    def collate_fn(self, batch): #этот метод можно реализовывать и отдельно,
      texts, labels = list(zip(*batch))
      padded_texts = pad_sequence(texts, batch_first=True).to(self.device)
      labels = torch.LongTensor(labels).to(self.device)
      return padded_texts, labels
```

```python
class MLP(nn.Module):

    def __init__(self, vocab_size, embedding_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.emb2h = nn.Linear(embedding_dim, 60)
        self.batchnorm1 = nn.BatchNorm1d(60)  # нормализация батча
        self.act1 = nn.ReLU()
        self.dropout = nn.Dropout(p=0.4)
        self.h2out = nn.Linear(60, 10) # 10 классов


    def forward(self, text):
        embedded = self.embedding(text)   # переводим последовательность индексов в последовательность эмбеддингов
        mean_emb = torch.mean(embedded, dim=1)
        hidden = self.emb2h(mean_emb)
        hidden = self.batchnorm1(hidden) # нормализация батча
        hidden = self.act1(hidden)
        hidden = self.dropout(hidden)
        out = self.h2out(hidden)
        return out
```

Перед обучением модели я открыла для себя такой чудесный параметр как patience.
Patience в нейронных сетях — это параметр, который указывает, через сколько эпох обучения без улучшения нужно прекратить процесс обучения. Например, если нейронная сеть уже 5 эпох подряд не улучшает свой результат (loss не уменьшается или accuracy не растёт), то, возможно, нет смысла учиться дальше.

```python
from tqdm import tqdm
from sklearn.metrics import classification_report, f1_score
from sklearn.utils import shuffle, class_weight
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
def train(model, iterator, optimizer, criterion):
    model.train()
    epoch_loss = 0
    progress_bar = tqdm(iterator, desc="Training", leave=False)  # добавила tqdm, чтобы отследить процесс обучения
    for i, (texts, ys) in enumerate(progress_bar):
        optimizer.zero_grad()
        outputs = model(texts)
        loss = criterion(outputs, ys)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
        progress_bar.set_postfix(loss=loss.item())
    return epoch_loss/len(iterator)
```

```python
def evaluate(model, iterator, criterion):
    model.eval()
    epoch_loss = 0
    all_preds = []
    all_true = []
    progress_bar = tqdm(iterator, desc="Evaluating", leave=False)
    with torch.no_grad():
        for i, (texts, ys) in enumerate(progress_bar):
            outputs = model(texts)
            loss = criterion(outputs, ys)
            epoch_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_true.extend(ys.cpu().numpy())
            progress_bar.set_postfix(loss=loss.item())
    #  F1 (weighted)
    f1 = f1_score(all_true, all_preds, average='weighted')
    return epoch_loss/len(iterator), f1
```

```python
def full_run(model, train_iterator, val_iterator, optimizer, criterion, num_epochs=50, patience=5, scheduler=None):
    train_losses = []
    val_losses = []
    val_f1s = []

    best_loss = np.inf
    epochs_no_improve = 0
    best_model_state = None

    for epoch in range(num_epochs):
        print(f"Epoch: {epoch+1}/{num_epochs}")
        tr_loss = train(model, train_iterator, optimizer, criterion)
        val_loss, val_f1 = evaluate(model, val_iterator, criterion)

        train_losses.append(tr_loss)
        val_losses.append(val_loss)
        val_f1s.append(val_f1)

        if scheduler is not None:
            scheduler.step()
            if epoch % 5 == 0:
                print(f'Epoch {epoch}, lr: {scheduler.get_last_lr()[0]}')

        if val_loss < best_loss:
            best_loss = val_loss
            best_model_state = model.state_dict()
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1

        if epoch % 5 == 0:
            print(f'Epoch {epoch}, train_loss={tr_loss:.4f}, val_loss={val_loss:.4f}, val_f1={val_f1:.4f}')

        # patience когда нет улучшения
        if epochs_no_improve >= patience:
            print(f'Early stopping at epoch {epoch}')
            break

    final_epoch = len(train_losses) - 1
    print(f'Final epoch: {final_epoch}, train_loss={train_losses[-1]:.4f}, val_loss={val_losses[-1]:.4f}, val_f1={val_f1s[-1]:.4f}')
    print(f'Best val_loss={best_loss:.4f}')

    # лучшие веса
    if best_model_state is not None:
        model.load_state_dict(best_model_state)

    # графики
    plt.figure(figsize=(10,4))
    plt.plot(train_losses, label='Train loss')
    plt.plot(val_losses, label='Val loss')
    plt.title('Loss')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10,4))
    plt.plot(val_f1s, label='Val F1')
    plt.title('Val F1')
    plt.legend()
    plt.show()
```

```python
class MLP(nn.Module):
    def __init__(self, vocab_size, embedding_dim=100): # каждое слово будет представлено вектором длиной 100
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)

        self.fc1 = nn.Linear(embedding_dim, 128) # увеличила с 60 до 128
        self.bn1 = nn.BatchNorm1d(128)
        self.act1 = nn.ReLU()
        self.dropout1 = nn.Dropout(p=0.5)

        self.fc2 = nn.Linear(128, 64)
        self.bn2 = nn.BatchNorm1d(64)
        self.act2 = nn.ReLU()
        self.dropout2 = nn.Dropout(p=0.3)

        self.out = nn.Linear(64, 10)

    def forward(self, text):
        embedded = self.embedding(text)
        mean_emb = torch.mean(embedded, dim=1)

        x = self.fc1(mean_emb)
        x = self.bn1(x)
        x = self.act1(x)
        x = self.dropout1(x)

        x = self.fc2(x)
        x = self.bn2(x)
        x = self.act2(x)
        x = self.dropout2(x)

        outputs = self.out(x)
        return outputs
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from torch.optim.lr_scheduler import ReduceLROnPlateau

def full_run(model, train_iterator, val_iterator, optimizer, criterion, num_epochs=50, patience=5, scheduler=None):
    train_losses = []
    val_losses = []
    val_f1s = []

    best_loss = np.inf
    epochs_no_improve = 0
    best_model_state = None

    for epoch in range(num_epochs):
        start_time = time.time()  # Начало отслеживания времени
        print(f"Epoch: {epoch+1}/{num_epochs}")

        tr_loss = train(model, train_iterator, optimizer, criterion)
        val_loss, val_f1 = evaluate(model, val_iterator, criterion)

        train_losses.append(tr_loss)
        val_losses.append(val_loss)
        val_f1s.append(val_f1)

        if scheduler is not None:
            scheduler.step(val_loss)  # Передаем валидационную потерю
            if epoch % 5 == 0:
                print(f'Epoch {epoch}, lr: {scheduler.get_last_lr()[0]}')

        if val_loss < best_loss:
            best_loss = val_loss
            best_model_state = model.state_dict()
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1

        if epoch % 5 == 0:
            print(f'Epoch {epoch}, train_loss={tr_loss:.4f}, val_loss={val_loss:.4f}, val_f1={val_f1:.4f}')

        if epochs_no_improve >= patience:
            print(f'Early stopping at epoch {epoch}')
            break

        end_time = time.time()  # Конец отслеживания времени
        print(f"Epoch {epoch+1} completed in {end_time - start_time:.2f} seconds.")

    final_epoch = len(train_losses) - 1
    print(f'Final epoch: {final_epoch}, train_loss={train_losses[-1]:.4f}, val_loss={val_losses[-1]:.4f}, val_f1={val_f1s[-1]:.4f}')
    print(f'Best val_loss={best_loss:.4f}')

    if best_model_state is not None:
        model.load_state_dict(best_model_state)

    plt.figure(figsize=(10,4))
    plt.plot(train_losses, label='Train loss')
    plt.plot(val_losses, label='Val loss')
    plt.title('Loss')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10,4))
    plt.plot(val_f1s, label='Val F1')
    plt.title('Val F1')
    plt.legend()
    plt.show()
```

```python
class MLP(nn.Module):
    def __init__(self, vocab_size, embedding_dim, pretrained_embeddings):
        super().__init__()
        self.embedding = nn.Embedding.from_pretrained(pretrained_embeddings, freeze=True)

        self.emb2h1 = nn.Linear(embedding_dim, 128)
        self.act1 = nn.LeakyReLU()
        self.dropout1 = nn.Dropout(p=0.2)

        self.emb2h2 = nn.Linear(128, 60)
        self.act2 = nn.LeakyReLU()
        self.dropout2 = nn.Dropout(p=0.2)

        self.h2out = nn.Linear(60, 10)

    def forward(self, text):
        embedded = self.embedding(text) # переводим последовательность индексов в последовательность эмбеддингов
        mean_emb = torch.mean(embedded, dim=1)
        hidden1 = self.emb2h1(mean_emb) # пропускаем эмбеддинг через полносвязный слой
        hidden1 = self.act1(hidden1)
        hidden1 = self.dropout1(hidden1)

        hidden2 = self.emb2h2(hidden1)
        hidden2 = self.act2(hidden2)
        hidden2 = self.dropout2(hidden2)

        out = self.h2out(hidden2) # возвращаем одно число для каждого семпла
        return out
```

```python
def full_run(model, train_iterator, val_iterator, optimizer, criterion, num_epochs=50, patience=5, scheduler=None):
    train_losses = []
    val_losses = []
    val_f1s = []

    best_loss = np.inf
    epochs_no_improve = 0
    best_model_state = None

    for epoch in range(num_epochs):
        print(f"Epoch: {epoch+1}/{num_epochs}")
        tr_loss = train(model, train_iterator, optimizer, criterion)
        val_loss, val_f1 = evaluate(model, val_iterator, criterion)

        train_losses.append(tr_loss)
        val_losses.append(val_loss)
        val_f1s.append(val_f1)

        if scheduler is not None:
            scheduler.step(val_loss)
            if epoch % 5 == 0:
                print(f'Epoch {epoch}, lr: {optimizer.param_groups[0]["lr"]}')


        if val_loss < best_loss:
            best_loss = val_loss
            best_model_state = model.state_dict()
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1

        if epoch % 5 == 0:
            print(f'Epoch {epoch}, train_loss={tr_loss:.4f}, val_loss={val_loss:.4f}, val_f1={val_f1:.4f}')

        # если несколько эпох подряд (patience) нет улучшения
        if epochs_no_improve >= patience:
            print(f'Early stopping at epoch {epoch}')
            break

    final_epoch = len(train_losses) - 1
    print(f'Final epoch: {final_epoch}, train_loss={train_losses[-1]:.4f}, val_loss={val_losses[-1]:.4f}, val_f1={val_f1s[-1]:.4f}')
    print(f'Best val_loss={best_loss:.4f}')

    # загружаем лучшие веса
    if best_model_state is not None:
        model.load_state_dict(best_model_state)

    # графики
    plt.figure(figsize=(10,4))
    plt.plot(train_losses, label='Train loss')
    plt.plot(val_losses, label='Val loss')
    plt.title('Loss')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10,4))
    plt.plot(val_f1s, label='Val F1')
    plt.title('Val F1')
    plt.legend()
    plt.show()

```
